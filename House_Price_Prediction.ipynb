```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Loading the dataset:-

url="https://raw.githubusercontent.com/diggajbishnoi/House-Price-Prediction/main/Bengaluru_House_Data.csv"
data = pd.read_csv(url)
data
```

```python
data.describe()
```

```python
## null values on the data set:-

data.isna().sum()
```

```python
## dropping the columns

data.drop(columns=["area_type","availability","society","balcony"],inplace=True)
```

```python
data.info()
```

```python
## counting the unique values:-

data["location"].value_counts()
```

```python
data["location"]=data["location"].fillna("Sarjapur Road")
```

```python
## unique values in size:-

data['size'].value_counts()
```

```python
## filling the missing values of the "Size" with 2 bhk:-

data['size']=data['size'].fillna('2BHK')
```

```python
## filling the missing values if the bathroom with their median:-

data['bath']=data['bath'].fillna(data['bath'].median())
```

```python
data.info()    ## now it shows that it has no null values.
```

```python
## converting the bedroom to bkh in size column:-

data['bhk'] = data['size'].str.extract('(\d+)').astype(float)
print(data)
```

```python
data[data.bhk > 20]    ## these are outliers in dataset. we have to fix them
```

```python
data['total_sqft'].unique()
```

```python
def convertRange(x):             ## this function is made to convert the range in the total sqft to a float number.

  temp=x.split('-')
  if len(temp)==2:
    return(float(temp[0])+float(temp[1]))/2
  try:
    return float(x)
  except:
    return None
```

```python
data['total_sqft']=data['total_sqft'].apply(convertRange)
```

```python
data.head()
```

```python
## Price per square feet:-

data['price_per_sqft']=data['price']*100000 / data['total_sqft']
```

```python
data['price_per_sqft']
```

```python
data.describe()
```

```python
data['location'].value_counts()
```

```python
data['location']=data['location'].apply(lambda x: x.strip())
location_count=data['location'].value_counts()
```

```python
location_count
```

```python
location_count_less_10=location_count[location_count<=10]
location_count_less_10
```

```python
data['location']=data['location'].apply(lambda x: 'other' if x in location_count_less_10 else x)
```

```python
data['location'].value_counts()
```

### Outlier detection and removal:-

```python
data.describe()
```

```python
(data['total_sqft']/data['bhk']).describe()
```

```python
data=data[((data['total_sqft']/data['bhk'])>=300)]
data.describe()
```

```python
data.shape
```

```python
data.price_per_sqft.describe()
```

```python
data.drop(columns=['size','price_per_sqft'],inplace=True)
data
```

### Cleaned Data

```python
data.head()
```

```python
data.to_csv("Cleaned_house_data.csv")
```

```python
X=data.drop(columns=['price'])
y=data['price']
```

### Machine Learning:-

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
print(X_train.shape)
print(y_train.shape)
```

### Applying Linear Regression

```python
column_trans = make_column_transformer((OneHotEncoder(sparse=False),['location']),
                                      remainder='passthrough')
```

```python
scaler=StandardScaler()
```

```python
lr = LinearRegression()
```

```python
pipe=make_pipeline(column_trans,scaler, lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_lr = pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred_lr)
```

### Applying Lasso

```python
lasso = Lasso()
```

```python
pipe = make_pipeline(column_trans,scaler, lasso)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_lasso = pipe.predict(X_test)
r2_score(y_test,y_pred_lasso)
```

### Applying Ridge

```python
ridge = Ridge()
```

```python
pipe = make_pipeline(column_trans,scaler, ridge)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_ridge = pipe.predict(X_test)
r2_score(y_test,y_pred_ridge)
```

```python
print("NO Regularization: ",r2_score(y_test,y_pred_lr))
print("Lasso: ",r2_score(y_test,y_pred_lasso))
print("Ridge: ",r2_score(y_test,y_pred_ridge))
```
